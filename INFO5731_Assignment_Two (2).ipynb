```python
# Write your code here

# #2 from IMDB

from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
from pathlib import Path
import os

url = 'https://www.imdb.com/title/tt1877830/reviews/?ref_=tt_ov_rt'

response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
#print(soup.text)
movies = soup.select('a.title')
movies2 = soup.select('span.review-date')
movies3 = soup.select('div.text.show-more__control')

print(len(movies))
print(len(movies2))
print(len(movies3))

titleList = []
dateList = []
descList = []


for index in range(0, len(movies)):
	titleList.append(movies[index].get_text())
print(titleList)

for index in range(0, len(movies2)):
	dateList.append(movies2[index].get_text())
print(dateList)

for index in range(0, len(movies3)):
	descList.append(movies3[index].get_text())
print(descList)

movie_review = pd.DataFrame({
    "Review": titleList,
    "Date": dateList,
    "Comment": descList
})
#movie_review.to_csv(index=False)
movie_review.to_csv('movies.csv')
```

```python
# importing required libraries

import pandas
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from textblob import Word
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
 
```

```python
# Write your code here

movies_df = pandas.read_csv('movies123.csv')
print(movies_df)
```

```python
#Remove noise, such as special characters and punctuations

updated_review_col = []
updated_comment_col = []

for i in movies_df['Review']:
    updated_review_col.append(re.sub("[^A-Za-z0-9 ]","",i))

movies_df['Review'] = updated_review_col

for i in movies_df['Comment']:
    updated_comment_col.append(re.sub("[^A-Za-z0-9 ]","",i))

movies_df['Comment'] = updated_comment_col

print(movies_df)
```

```python
#Remove numbers.

updated_review_col = []
updated_comment_col = []

for i in movies_df['Review']:
    updated_review_col.append(re.sub("[^A-Za-z ]","",i))

movies_df['Review'] = updated_review_col

for i in movies_df['Comment']:
    updated_comment_col.append(re.sub("[^A-Za-z ]","",i))

movies_df['Comment'] = updated_comment_col

print(movies_df)
```

```python
#Lowercase all texts

updated_review_col = []
updated_comment_col = []

for i in movies_df['Review']:
    updated_review_col.append(i.lower())

movies_df['Review'] = updated_review_col

for i in movies_df['Comment']:
    updated_comment_col.append(i.lower())

movies_df['Comment'] = updated_comment_col

print(movies_df)
```

```python
# Removing stopwords

stop_words = set(stopwords.words('english'))

print('Comments column after removing stopwords:')
filtered_column_comment = []
for i in movies_df['Comment']:
    word_tokens = word_tokenize(i)
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
    filtered_sentence = []
    
    for w in word_tokens:
        if w not in stop_words:
            filtered_sentence.append(w) 
    print(filtered_sentence)
    filtered_column_comment.append(filtered_sentence)
print(filtered_column_comment)
```

```python
# Removing stopwords

print('Reviews column after removing stopwords:')
filtered_column_review = []
for i in movies_df['Review']:
    word_tokens = word_tokenize(i)
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
    filtered_sentence = []
    
    for w in word_tokens:
        if w not in stop_words:
            filtered_sentence.append(w)  
    print(filtered_sentence)
    filtered_column_review.append(filtered_sentence)
    
#print(filtered_column_review)
```

```python
# stemming

print('After applying Stemming to Review column')

portstem = PorterStemmer()

for i in filtered_column_review:
    updated_review_column_stem = []
    words = word_tokenize(" ".join(i))
    for w in words:
        updated_review_column_stem.append(portstem.stem(w))
    print(updated_review_column_stem)
```

```python
# stemming

print('After applying Stemming to Comments column')

portstem = PorterStemmer()

for i in filtered_column_comment:
    updated_comment_column_stem = []
    words = word_tokenize(" ".join(i))
    for w in words:
        updated_comment_column_stem.append(portstem.stem(w))
    print(updated_comment_column_stem)
```
