```python
# (3) Extract all the noun phrases and calculate the relative probabilities of each review in terms of other reviews (abstracts, or tweets) by using the fomular frequency (noun phrase) 
# / max frequency (noun phrase) on the whole dataset. Print out the result in a table with column name the all the noun phrases and row name as all the 100 reviews (abstracts, or tweets).


# importing required libraries, from nltk for NLP
import nltk
from nltk import word_tokenize, pos_tag
nltk.download('averaged_perceptron_tagger')

# find the parts of speech for each tokens/words
parts_of_speech = nltk.pos_tag(tokens)

# filter out the terms which are nouns
nouns = list(filter(lambda x: x[1] == "NN", parts_of_speech))

# print the list of nouns
nouns

# convert the list into nouns list
li = []
for n in nouns:
  li.append(n[0])

# find unique terms by converting into set
se = set(li)

# create dictionary
dic = {}

# create dictionary containing the noun term & its frequency
for s in se:
  count = li.count(s)
  dic[s] = count


# retrieve all the frequency values using dictionary values
count_freq = dic.values()

# calculate the total of frequencies
total = 0
for c in count_freq:
  total = total + c

# calculate the answer by using the formula fomular frequency (noun phrase) / max frequency (noun phrase)
for d in dic:
  dic[d] = dic[d]/total

# print the answer using dataframe
import pandas as pd
df = pd.DataFrame(dic, index = ['relative probability'])
print(df)
```

# **Question 2: Undersand TF-IDF and Document representation**

(20 points). Starting from the documents (all the reviews, or abstracts, or tweets) collected for assignment two, write a python program: 

(1) To build the **documents-terms weights (tf*idf) matrix bold text**.

(2) To rank the documents with respect to query (design a query by yourself, for example, "An Outstanding movie with a haunting performance and best character development") by using **cosine similarity**.

```python
from pandas.core.arrays.timedeltas import TD64NS_DTYPE
# Write your code here
import numpy as np

# (1) To build the documents-terms weights (tf*idf) matrix bold text.

#Preprocessing the text data

# creating empty lists to hold the processed data
s = []
ws = []


# run the for loop for 'Comment' column of the csv file
for sent in df_movies['Comment']:
    # convert the words into lower case and append into the list sentences
    x = [i.lower() for  i in word_tokenize(sent) if i.isalpha()]
    s.append(x)
    # append the words into word_set to finally convert into set and find the unique words
    for word in x:
        if word not in ws:
            ws.append(word)
 
# create a set of vocabulary
ws = set(ws)
# find the no. of documents
td = len(s)
 
# using indexes for each word in the documents list
index_dict = {} 
i = 0
for word in ws:
    # create a dictionary of words, along with indexes
    index_dict[word] = i
    i += 1

# create a dictionary of words, along with counts, with words saved as key
def find_count(s):
    wc = {}
    for word in ws:
        wc[word] = 0
        for sent in s:
            if word in sent:
                wc[word] += 1
    return wc

# create the word count dictionary using the function just created
wc = find_count(s)

# find the TF - Term frequency
def termfreq(document, word):
    # find the document length
    N = len(document)
    # find the occurence of the word tokens
    occurance = len([token for token in document if token == word])
    return occurance/N


# find the IDF score
def inverse_doc_freq(word):
    try:
        word_occurance = wc[word] + 1
    except:
        word_occurance = 1
    return np.log(td/word_occurance)

# create a function to find the TF-IDF values and vector
def tf_idf(sentence):
    # create a TF-IDF vector
    tf_idf_vec = np.zeros((len(ws),))
    for word in sentence:
        # find the Term frequency TF
        tf = termfreq(sentence,word)
        # find the IDF score
        idf = inverse_doc_freq(word)
        # find the TF-IDF value
        value = tf*idf
        # create and return the TF-IDF vector
        tf_idf_vec[index_dict[word]] = value 
    return tf_idf_vec
```

```python
#To rank the documents with respect to query (design a query by yourself, for example, "An Outstanding movie with a haunting performance and best character development") by using cosine similarity.

# importing the required libraries
from re import sub
from gensim.utils import simple_preprocess

# created a sample query string to compare
query_string = 'batman movie is a great movie'
# set the created list containing entire word list into d variable
ds = lst
# create a custom list of stopwords and added some common words into the same list
stopwords = ['the', 'and', 'are', 'a']
```

```python
# define a function to preprocess and clean the data
def preprocess(doc):
    # below step helps to clean the data and return the tokenized version
    doc = sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', " url_token ", doc)
    # return the token values removing the stopwords
    return [token for token in simple_preprocess(doc, min_len=0, max_len=float("inf")) if token not in stopwords]
```

```python
# Preprocess the documents, including the query string
# create the corpus using documents by preprocessing function defined above
corpus = [preprocess(d) for d in ds]
```

```python
# preprocess the query string defined above to convert into cleaned and tokenized version
query = preprocess(query_string)
```

```python
# import the required libraries in gensim for cosine similarity function
import gensim.downloader as api
from gensim.corpora import Dictionary
from gensim.models import TfidfModel
from gensim.similarities import WordEmbeddingSimilarityIndex
from gensim.similarities import SparseTermSimilarityMatrix
from gensim.similarities import SoftCosineSimilarity

# use the glove model to determine cosine similarity
g = api.load("glove-wiki-gigaword-50")    
# determine the similarity index
similarity_index = WordEmbeddingSimilarityIndex(g)
# using TFIDF model, create a dictionary
d = Dictionary(corpus+[query])
tf = TfidfModel(d=d)
# creation of the similarity matrix
sm = SparseTermSimilarityMatrix(similarity_index, d, tf)
```

```python
# import the required libraries
import numpy as np

#create the TF-IDF query using doc2bow function on the query
qt = tf[d.doc2bow(query)]
# create index value using cosine similarity function
i = SoftCosineSimilarity(tf[[d.doc2bow(document) for document in corpus]],sm)
ss = i[qt]

# display the cosine similarities between entered query and the comment values from the csv file
si = np.argsort(ss)[::-1]
for idx in si:
    # print the results
    print(f'{idx} \t {ss[idx]:0.3f} \t {ds[idx]}')
```

```python
# import the data from csv
p = "movies.csv"
# import data into dataframe
df = pd.read_csv(p)
```

```python
# cleaning & preprocessing functions to clean and lowercase the data entered
def cl(t):
    t = re.sub(r'[^ \nA-Za-z0-9À-ÖØ-öø-ÿ/]+', '', t)
    t = re.sub(r'[\\/×\^\]\[÷]', '', t)
    return t
def lower(t):
    t = t.lower()
    return t
```

```python
def buildmodel(df):
    df2 = pd.DataFrame(df["Comment"]).values.tolist()
    for i in range(len(df2)):
        df2[i] = df2[i][0].split(" ")
    print(df2)
    return df2
```

```python
def trainmodel(df2):
    cores = multiprocessing.cpu_count()
    model2 = Word2Vec(min_count=4, window=4, alpha=0.03, min_alpha=0.0007, sg = 1, workers=cores-1)
    model2.build_vocab(df2, progress_per=10000)
    model2.train(df2, total_examples=model2.corpus_count, epochs=100, report_delay=1)
    return model2
```

```python
# apply the functions created above
df[["Comment"]] = df[["Comment"]].astype(str)
df["Comment"] = df["Comment"].apply(lower)
df["Comment"] = df["Comment"].apply(cl)
```

```python
# build the model
df3 = buildmodel(df)
```
