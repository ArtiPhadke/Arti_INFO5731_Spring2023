```python
# read the first 1000 rows data from netflix_titles.csv file
movies_df = pandas.read_csv('netflix_titles.csv',nrows = 1000)

# display few rows from the dataframe
print(movies_df)

# columns of interest are: title and description, which are text columns & can help extract the features of interest
```

```python
print('Convert the text in Description column to lowercase:')
# 
description = []

for w in movies_df.description:
  #convert to lowercase
  description.append(w.lower())

# print the lowercase description
print(description)

print('Convert the text in title column to lowercase:')
titles = []

for w in movies_df.title:
  #convert to lowercase
  titles.append(w.lower())

# print the lowercase titles
print(titles)
```

```python
# code to remove the stopwords

from nltk.classify.maxent import train_maxent_classifier_with_gis
stop_words = set(stopwords.words('english'))

description2 = []

# code to remove the stopwords from description column

print('Remove the stopwords from description column')
for i in description:
    # taking out the word tokens from the description column
    word_tokens = word_tokenize(i)
    # Check if the word token is part of the stop_words
    description2 = [w for w in word_tokens if not w.lower() in stop_words]
    description2 = []
    
    for w in word_tokens:
        if w not in stop_words:
            # if not in stop_words, then append to the updated list
            description2.append(w) 
    # display the updated description
    print(description2)


title2 = []

# code to remove the stopwords from title column

print('Remove the stopwords from title column')
for i in titles:
    # taking out the word tokens from the title column
    word_tokens = word_tokenize(i)
    # Check if the word token is part of the stop_words
    title2 = [w for w in word_tokens if not w.lower() in stop_words]
    title2 = []
    
    for w in word_tokens:
        if w not in stop_words:
            # if not in stop_words, then append to the updated list
            title2.append(w) 
    # display the updated title
    print(title2)
```

```python
# stemming

print('After applying Stemming to Description column')

# import the function to apply stemming
portstem = PorterStemmer()

# run the loop for each value from Review column
for i in description2:
    description3 = []
    # take out the word tokens and append those
    words = word_tokenize(" ".join(i))
    for w in words:
        # create updated column data, after running the word tokens through porter stemmer and stemming process
        description3.append(portstem.stem(w))
    # display each comment after stemming
    print(description3)
```

```python
# stemming

print('After applying Stemming to Titles column')

# import the function to apply stemming
portstem = PorterStemmer()

# run the loop for each value from Review column
for i in title2:
    title3 = []
    # take out the word tokens and append those
    words = word_tokenize(" ".join(i))
    for w in words:
        # create updated column data, after running the word tokens through porter stemmer and stemming process
        title3.append(portstem.stem(w))
    # display each comment after stemming
    print(title3)
```

```python
# (1) Parts of Speech (POS) Tagging: Tag Parts of Speech of each word in the text, and calculate the total number of N(oun), V(erb), Adj(ective), Adv(erb), respectively.

# Parts of Speech (POS) from 'Description' column

# importing required libraries for POS
import spacy
nlp=spacy.load('en_core_web_sm')

# create variables to hold each POS count, N(oun), V(erb), Adj(ective), Adv(erb)
verb_count = 0
Adjective_count = 0
Noun_count = 0
Adverb_count = 0

# running the loop to select each sentence from 'Description' column, using cleaned data after removing stopwords
for i in description:
  # converting list to String
  str = " ".join(i)
  # for each selected word from str, find the text, and Part of Speech
  for word in nlp(str):
    #print(word.text, ':',word.pos_)
    # if given word is Noun, increase Noun count
    if(word.pos_ == 'NOUN'):
      Noun_count+=1
    # if given word is Adj, increase Adj count
    elif(word.pos_ == 'ADJ'):
      Adjective_count+=1  
    # if given word is Adv, increase Adv count
    elif(word.pos_ == 'ADV'):
      Adverb_count+=1
    # if given word is Verb, increase Verb count
    elif(word.pos_== 'VERB'):
      verb_count+=1

# print the no. of N(oun), V(erb), Adj(ective), Adv(erb)
print('No. of Adverbs = ',Adverb_count)
print('No. of Verbs = ',verb_count)
print('No. of Nouns = ',Noun_count)
print('No. of Adjectives = ',Adjective_count)      
```
